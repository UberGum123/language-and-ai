# Data cleaning

```python
# import libraries
import pandas as pd

import re
import unicodedata
import contractions

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
from nltk.probability import FreqDist
```

```python
# Load the data
df = pd.read_csv('../data/political_leaning.csv')

text = df['post'][0]
text
```

```python
def remove_formatting(text):
    text = re.sub(r'\s+', ' ', text)  # collapse whitespace
    return text.strip()

text_clean = remove_formatting(text)
print(text_clean)
```

```python
def expand_contractions(text):
    expanded_text = contractions.fix(text)
    return expanded_text

text_expanded = expand_contractions(text_clean)
print(text_expanded)
print(len(text_expanded))
```

```python
tokens = word_tokenize(text_expanded)
print(tokens)
```

```python
tokens = [t.lower() for t in tokens if t.isalpha()]
print(tokens)
print(len(tokens))
```

```python
stop_words = set(stopwords.words('english'))

tokens = [t for t in tokens if t not in stop_words]
print(tokens)
print(len(tokens))
```

```python
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    if tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

pos_tags = pos_tag(tokens)

lemmas = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in pos_tags
]

print(lemmas)
```

```python
# Check for bigrams (Maybe not necessary tbh)
finder = BigramCollocationFinder.from_words(lemmas)
finder.apply_freq_filter(2)  # frequency cutoff

bigrams = finder.nbest(BigramAssocMeasures.pmi, 10)
print(bigrams)
```
